```python
%%manim -qm -v WARNING SquareToCircle

class SquareToCircle(Scene):
   def construct(self):
      square = Square()
      circle = Circle()
      circle.set_fill(PINK, opacity=0.5)
      self.play(Create(square))
      self.play(Transform(square, circle))
      self.wait()
```

```python
%%manim -qm -v WARNING MatrixDisplay
class MatrixDisplay(Scene):
    def construct(self):
        title = Text("Matrix representation", font_size=36).to_edge(UP)
        A = Matrix([[2, 3],
                    [4, -1]], left_bracket="[" , right_bracket="]")
        A.scale(1.25)
        self.play(Write(title), Create(A))
        self.wait()
```

```python
%%manim -qm -v WARNING MatrixAdditionAnimation
class MatrixAdditionAnimation(Scene):
    def construct(self):
        # Define matrices
        A = Matrix([[1, 2], [3, 4]])
        B = Matrix([[5, 6], [7, 8]])
        C = Matrix([[6, 8], [10, 12]])

        # Position matrices
        A.move_to(LEFT * 4)
        B.move_to(RIGHT * 4)
        C.move_to(ORIGIN)

        # Display matrices A and B
        self.play(Write(A), Write(B))
        self.wait(1)

        # Highlight and add elements step-by-step
        for i in range(2):
            for j in range(2):
                # Get the matrix entries (MathTex objects)
                a_entry = A.get_entries()[i*2+j]
                b_entry = B.get_entries()[i*2+j]
                c_entry = C.get_entries()[i*2+j]

                # Highlight A[i,j] and B[i,j]
                self.play(
                    a_entry.animate.set_color(YELLOW),
                    b_entry.animate.set_color(YELLOW)
                )
                # Show addition with Tex, using tex_string for safety
                addition_text = Tex(
                    f"${a_entry.tex_string} + {b_entry.tex_string} = {c_entry.tex_string}$"
                ).next_to(C, UP)
                self.play(Write(addition_text))
                # Fill in C[i,j]
                self.play(Write(c_entry))
                # Fade out the addition text
                self.play(FadeOut(addition_text))
                self.wait(1)

        # Highlight the final result
        self.play(Indicate(C))
        self.wait(2)
```

```python
%%manim -qm -v WARNING MatrixSubtractionAnimation

class MatrixSubtractionAnimation(Scene):
    def construct(self):
        # Define matrices
        A = Matrix([[5, 6], [7, 8]])
        B = Matrix([[1, 2], [3, 4]])
        C = Matrix([[4, 4], [4, 4]])

        # Position matrices
        A.move_to(LEFT * 4)
        B.move_to(RIGHT * 4)
        C.move_to(ORIGIN)

        # Display matrices A and B
        self.play(Write(A), Write(B))
        self.wait(1)

        # Highlight and subtract elements step-by-step
        for i in range(2):
            for j in range(2):
                # Get the matrix entries (MathTex objects)
                a_entry = A.get_entries()[i*2+j]
                b_entry = B.get_entries()[i*2+j]
                c_entry = C.get_entries()[i*2+j]

                # Highlight A[i,j] and B[i,j]
                self.play(
                    a_entry.animate.set_color(YELLOW),
                    b_entry.animate.set_color(YELLOW)
                )
                # Show subtraction with Tex, using tex_string for safety
                subtraction_text = Tex(
                    f"${a_entry.tex_string} - {b_entry.tex_string} = {c_entry.tex_string}$"
                ).next_to(C, UP)
                self.play(Write(subtraction_text))
                # Fill in C[i,j]
                self.play(Write(c_entry))
                # Fade out the subtraction text
                self.play(FadeOut(subtraction_text))
                self.wait(1)

        # Highlight the final result
        self.play(Indicate(C))
        self.wait(2)
```

```python
%%manim -qm -v WARNING MatrixSubtractionAnimation

class MatrixMultiplicationAnimation(Scene):
    def construct(self):
        # Define matrices
        A = Matrix([[1, 2], [3, 4]])
        B = Matrix([[5, 6], [7, 8]])
        C = Matrix([[19, 22], [43, 50]])

        # Position matrices
        A.move_to(LEFT * 4)
        B.move_to(RIGHT * 4)
        C.move_to(ORIGIN)

        # Display matrices A and B
        self.play(Write(A), Write(B))
        self.wait(1)

        # Highlight and compute elements step-by-step
        for i in range(2):
            for j in range(2):
                # Get the matrix entries (MathTex objects)
                c_entry = C.get_entries()[i*2+j]

                # Highlight row i of A and column j of B
                row_a = A.get_entries()[i*2:i*2+2]  # Entries in row i
                col_b = [B.get_entries()[j], B.get_entries()[j+2]]  # Entries in column j

                self.play(
                    *[entry.animate.set_color(YELLOW) for entry in row_a],
                    *[entry.animate.set_color(YELLOW) for entry in col_b]
                )

                # Show multiplication and summation
                # Example for C_{11}: 1*5 + 2*7 = 19
                terms = [
                    f"{A.get_entries()[i*2+k].tex_string} \\times {B.get_entries()[k*2+j].tex_string}"
                    for k in range(2)
                ]
                summation_text = Tex(
                    f"${' + '.join(terms)} = {c_entry.tex_string}$"
                ).next_to(C, UP)
                self.play(Write(summation_text))

                # Fill in C[i,j]
                self.play(Write(c_entry))

                # Fade out the summation text and reset colors
                self.play(
                    FadeOut(summation_text),
                    *[entry.animate.set_color(WHITE) for entry in row_a],
                    *[entry.animate.set_color(WHITE) for entry in col_b]
                )
                self.wait(1)

        # Highlight the final result
        self.play(Indicate(C))
        self.wait(2)
```
